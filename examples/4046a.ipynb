```python
from pymna import Circuit
from pymna.units import kOhm,uF, V, Hz
from pymna.simulator import Simulator
import matplotlib.pyplot as plt
import os

%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%matplotlib ipympl
%autoreload 2
basepath = os.getcwd()
print(basepath)
```

```python
def get_reference( path ):
    with open(path, 'r') as f:
        lines = f.readlines()
        keys = lines[0].strip().split()
        d={k:[] for k in keys}
        print(keys)
        for l in lines[1::]:
            ll=l.strip().split()
            for idx, key in enumerate(keys):
                d[key].append( float(ll[idx]) )
        return d

ref = get_reference( f'{basepath}/../mna/references/4046a_Simulated.tab')
```

```python
path = f"{basepath}/../mna/netlists/4046a.net"
simulator = Simulator()
results = simulator.run_from_nl(path)


```

```python
plt.style.use('dark_background')

fig, ax = plt.subplots()
plt.plot( results['t'], results['5'] )
#plt.plot( ref['t'], ref['5']  )

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title("")

plt.show()
```
