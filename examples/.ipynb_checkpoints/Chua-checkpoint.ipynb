```python
import pymna
import numpy as np
from pymna import Circuit, Simulator
from pymna import elements as elm
from pymna.units import kOhm,uF, V, Hz
import matplotlib.pyplot as plt
import os


module_path = pymna.__file__
basepath = os.path.dirname(module_path)
print(basepath)

%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%matplotlib ipympl
%autoreload 2
```

```python
def get_reference( path ):
    with open(path, 'r') as f:
        lines = f.readlines()
        keys = lines[0].strip().split()
        d={k:[] for k in keys}
        print(keys)
        for l in lines[1::]:
            ll=l.strip().split()
            for idx, key in enumerate(keys):
                d[key].append( float(ll[idx]) )
        return d

def get_distance(target, a):
    return np.abs(np.array(target)-np.array(a)).sum()
    

```

## Chua circuit:

```python
ref = get_reference( f'{basepath}/tests/data/results/Chua.sim')
path = f"{basepath}/tests/data/netlists/Chua.net"
simulator = Simulator()
circuit, results = simulator.run_from_nl(path)
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots()
plt.plot( results['1'], results['2'] )
plt.plot( ref['1'], ref['2']  )
ax.set_xlabel('Node 1 (V)')
ax.set_ylabel('Node 2 (V)')
ax.set_title("Chua")
plt.show()
plt.savefig("x.png",dpi=1000)
```

```python
for key in results.keys():
    print(get_distance(ref[key] , results[key]))
```
